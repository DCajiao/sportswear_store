```python
import pymysql
import pandas as pd

def conectar_db(host, usuario, contraseña, db):
    try:
        conexion = pymysql.connect(host=host,
                                    user=usuario,
                                    password=contraseña,
                                    database=db)
        print("Conexión exitosa a la base de datos")
        return conexion
    except Exception as e:
        print("Error al conectar a la base de datos:", e)
        return None

def consultar_tablas(conexion):
    try:
        with conexion.cursor() as cursor:
            cursor.execute("SHOW TABLES;")
            tablas = cursor.fetchall()
            return [tabla[0] for tabla in tablas]
    except Exception as e:
        print("Error al consultar las tablas:", e)
        return None

def consultar_vista(conexion, vista):
    try:
        with conexion.cursor() as cursor:
            cursor.execute(f"SELECT * FROM {vista};")
            resultados = cursor.fetchall()
            columnas = [desc[0] for desc in cursor.description]
            return pd.DataFrame(resultados, columns=columnas)
    except Exception as e:
        print(f"Error al consultar la vista {vista}:", e)
        return None

def traer_datos_tabla(conexion, tabla):
    try:
        with conexion.cursor() as cursor:
            cursor.execute(f"SELECT * FROM {tabla};")
            resultados = cursor.fetchall()
            columnas = [desc[0] for desc in cursor.description]
            return pd.DataFrame(resultados, columns=columnas)
    except Exception as e:
        print(f"Error al traer los datos de la tabla {tabla}:", e)
        return None

def main():
    global df_top3productosmasvendidos
    # Configuración de la conexión a la base de datos
    host = "localhost" 
    usuario = "root"
    contraseña = "root"
    db = "sportswear_store"

    # Conexión a la base de datos
    conexion = conectar_db(host, usuario, contraseña, db)
    if conexion is None:
        return

    # Consultar las tablas de la base de datos
    tablas = consultar_tablas(conexion)
    if tablas is None:
        conexion.close()
        return

    print("\nTablas en la base de datos:")
    for tabla in tablas:
        print(tabla)

    # Consultar y mostrar los datos de una vista específica
    vista = "top3productosmasvendidos"
    df_top3productosmasvendidos = consultar_vista(conexion, vista)
    if df_top3productosmasvendidos is not None:
        print("\nDatos de la vista:", vista)
        print(df_top3productosmasvendidos)

    conexion.close()

if __name__ == "__main__":
    main()
```
